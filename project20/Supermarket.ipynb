```python
plt.figure(figsize=(8,8))
plt.style.use('default')
ax= sns.countplot('Gender', data=df , palette = 'crest')
ax.set_xlabel(xlabel= "Gender",fontsize=18)
ax.set_ylabel(ylabel = "Gender count", fontsize = 10)
ax.set_title(label = "Cinsiyet ayrm", fontsize =15)
plt.show()
```

```python
plt.figure(figsize= (14,6))
ax = sns.countplot(x = "Customer type", data = df, palette = "flare")
ax.set_title("Mteri analizi", fontsize = 25)
ax.set_xlabel("Customer type", fontsize = 16)
ax.set_ylabel("Customer Count", fontsize = 16)
```

```python
plt.figure(figsize = (14,6))
ax = sns.countplot(x = "Payment", data = df, palette = "viridis")
ax.set_title(label = "deme Yntemleri ", fontsize= 25)
ax.set_xlabel(xlabel = "Payment method", fontsize = 16)
ax.set_ylabel(ylabel = " Customer Count", fontsize = 16)
```

```python
df["Time"]= pd.to_datetime(df["Time"])
df["Hour"]= (df["Time"]).dt.hour
```

```python
plt.figure(figsize=(14,6)) 
SalesTime = sns.lineplot(x="Hour", y ="Quantity", data = df).set_title("Saat Ba Satlar")
```

```python
sns.countplot(df['Quantity'])
```

```python
plt.figure(figsize=(10,6)) 
ax = sns.boxenplot(x = "Quantity", y = "Product line", data = df, scale="linear")
ax.set_title(label = "", fontsize = 25)
ax.set_xlabel(xlabel = "Quantity Sales",fontsize = 16)
ax.set_ylabel(ylabel = "Product Line", fontsize = 16)
```

```python
plt.figure(figsize=(14,6))
ax = sns.countplot(y='Product line', data=df, order = df['Product line'].value_counts().index)
ax.set_title(label = "rnlerin sat", fontsize = 25)
ax.set_xlabel(xlabel = "Sales count", fontsize = 16)
ax.set_ylabel(ylabel= "Product Line", fontsize = 16)
```

```python
plt.figure(figsize = (14,6))
ax= sns.stripplot(y= "Product line", x = "Total", hue = "Gender", data = df)
ax.set_title(label = "Cinsiyete Gre Satlar")
ax.set_xlabel(xlabel = " Total sales of products")
ax.set_ylabel(ylabel = "Product Line")
```
